```python
#For clarity
df = df.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish",
                   "MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
columns_to_drop = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 
                   'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp4', 
                   'Complain', 'Response', 'Kidhome', 'Teenhome']

df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
s = (df.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables in the dataset:", object_cols)

from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for i in object_cols:
    df[i]=df[[i]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df)
scaled_df = pd.DataFrame(scaler.transform(df),columns= df.columns )
print("All features are now scaled")
```
